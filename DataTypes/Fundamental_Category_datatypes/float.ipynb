float:

'float' is immutable type.

'float' is one of the pre-defined class name and treated as fundamental data type.

The purpose of float data type is to store decimal values or floating point numbers.

float data type never allows us to store/represent binary, octal and hexadecimal values.

float data types allows us to store Scientific Notation data.

```python
a=13.45
print(a,type(a))
```

```python
# Float Immutability

a=12.5
print(a,id(a))

a=a+1
print(a,id(a))
```

```python
# Float Object Reuse

a=1.00
print(a,type(a),id(a))

b=1.00
print(b,type(b),id(b))
```

```python
# Reference Assignment

a=2.34
b=a
print(a,type(a),id(a))
print(b,type(b),id(b))
```

```python
#  Float precision problem

print(0.1+0.2)

# Float values are stored internally in binary format, so some decimal values cannot be represented exactly.
```

```python
0.1+0.2
```

```python
# Scientific Notation

a=3e2
print(a,id(a),type(a))

b=3e-2
print(b,id(b),type(b))
```

```python
# Type Casting to float

print(float(89))
print(float(87.99))
print(float(True))
print(float("90"))
print(float("89.76"))
print(float("3e2"))
```

```python
# is vs ==
a=12.34
b=12.34

print(a==b)                 
print(a is b)               
```

```python
# Special float values

a=float('inf')
b=float('-inf')
c=float('nan')
print(a)
print(b)
print(c)
```
