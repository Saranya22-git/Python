str

'str' is one of the predefined class and treated as Sequence data type.

The purpose of str data type is "To store string data or text data or alphanumeric data or numeric data or special symbols within double quotes or single quotes or tripple double quotes or tripple single quotes."

```python
name="Python"
print(name,type(name))
name
```

```python
# Single line string data

name="Saranya"
branch='Ai&ds'

print(name)
print(branch)
```

```python
# Multiple Line String data

name="""Hello, Saranya. 
        How are you?"""

branch='''Artificial Intelligence and
          Data Science'''


print(name)
print(branch)
```

```python
# Immutability type

name="Python"

print(name,id(name))

name="Python"+'123'

print(name,id(name))
```

```python
# Indexing

# Example:1

s="Python"

print(s,type(s))
print(s[0])
print(s[-6])
print(s[5])
s[2]

print(s[-7])
```

```python
s="1234567"

print(len(s))
print(s[len(s)-1])
print(s[len(s)])
```

```python
# Slicing

# Syntax-1: [BEGIN:END]

s="PYTHON"

print(s,type(s))
print(s[0:3])
print(s[2:6])
print(s[1:6])
print(s[0:6])
s[2:1]
```
