What is a variable in Python?

A **variable** is a name that refers to a value stored in memory.

`x=10` Here x is a **variable name** and 10 is **value**

Does Python require variable declaration?

No, it is **dynamic typing**(Type is decided at runtime)

What are variable naming rules?

`name="sara"`

`x_=0099`

`_x=00990`

`name12="sara"`

- Special Symbols are not allowed except UNDERSCORE(_)

    `sara@="saranya"` - shows syntx error

    `sara n=34` - shows syntx error

- Cannot use Keywords as Variable Names

    `if =90` - shows syntx error

Is Python case-sensitive?

- Case-Sensitive
  
    `Age=45`

    `age=45`

```python
# What is variable unpacking?

# Taking values from a list/tuple and assigning them to variables in one line. Number of variables = number of values (must match)

a, b = [1, 2]

a, *b = [1, 2, 3, 4]

print(a)
print(b)
```
