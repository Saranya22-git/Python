Input(message)

Syntax=varname=input(message)

The input(message) is used for reading any type of data from keyboard in the form of str and additionally it will prompt user messages.

```python
''' Write a pyhton program which will calculate area of circle 

 Area of Circle=pi*r*r

'''

r=float(input("Enter the radius"))
ac=3.14*r*r

print("Radius={}".format(r))
print("Area of circle:{}".format(ac))

print("Area of circle:%0.2f" %ac)
```

```python
''' Write a python program which will calculate the perimeter of circle

Perimeter of Circle=2*pi*r

'''

r=float(input("Enter the radius:"))
pc=2*3.14*r

print("Radius={}".format(r))
print("Perimeter of Circle={}".format(pc))

print("Perimeter of Circle=%0.2f" %pc)
```

```python
''' Write a python program wich will calculate area and perimeter of rectangle

Area of rectangle=l*b
Perimeter of rectangle=2(l+b)

'''

l=float(input("Enter the length:"))
b=float(input("Enter the breadth:"))

ar=l*b
pr=2*(l+b)

print("Area of rectangle({},{})={}".format(l,b,ar))
print("Perimeter of rectangle({},{})={}".format(l,b,pr))
```

```python
''' Write a python program which will calculate area of triangle

Area of Triangle=(1/2)*b*h

'''

b=float(input("Enter the base:"))
h=float(input("Enter the height:"))

at=(1/2)*b*h

print("Area of triangle({},{})={}".format(b,h,at))
```

```python
''' Write a python program which will convert celsius temperature into fahrenheit and kelvin

f=(9/5)*c+32
k=c+273

'''

c=float(input("Enter the celsisu value:"))

f=(9/5)*c+32
k=c+273

print("Fahrenheit temperature={}".format(f))
print("Kelvin:{}".format(k))
```

```python
''' Write a python program which will calculate area and perimeter of square

Area of Square=s*s
Perimeter of Square=4*s

'''

s=float(input("Enter the side:"))

sa=s*s 
ps=4*s

print("Area of Square({},{})={}".format(s,s,sa))
print("Perimeter of Square({},{})".format(s,s,ps))
```
