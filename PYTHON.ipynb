```python
#-------SLICING--------

#-------SYNTAX-1-------
# strobj=[BEGIN:END]---BEGIN<END
s="Python"
print(s[0:2])
print("--------")
print(s[2:0]) # Here begin is greater than end so that's why we never get any result
print("--------")
print(s[0:5])
print("--------")
print(s[-6:-1])
print("--------")
print(s[-1:-6])
print("--------")
print(s[2:7])
print("--------")
print(s[-10:-1])
print("--------")
print(s[:-6])
print("--------")
```

```python
#-------SYNTAX-2-------
# strobj=[BEGIN:]

s="saranya"
print(s[0:])
print("--------")
print(s[:-7])
print("--------")
print(s[3:])
print("--------")
print(s[-7:])
print("--------")
print(s[-678:])
print("--------")
print(s[45:])
print("--------")
```

```python
#-------SYNTAX-3-------
# strobj=[:END]

a="apple"
print(a[:-9000])
print("---------")
print(a[:-4])
print("---------")
print(a[:-2])
print("--------")
print(a[:-1])
print("--------")
```

```python
#-------SYNTAX-4-------
# strobj=[:]

b="balloon"
print(b[:])
print("--------")
print(b[:len(b)])
print("--------")
print(b[-len(b):])
print("--------")
```

```python
#-------SYNTAX-5-------
# strobj=[BEGIN:END:STEP]

# RULE-1: Here begin, end, step values are either be +ve or -ve

# RULE:2: If step value is +ve then it takes from forward direction provided by BEGIN INDEX<END INDEX otherwise we get space or '' as a result.

e="elephant"
print(e[0:3:1])
print("----------")
print(e[::1])
print("----------")
print(e[::2])
print("----------")
print(e[::3])
print("----------")
print(e[::5])
print("----------")
print(e[2:1:1])
print("----------")
```

```python
#-------SYNTAX-5-------
# strobj=[BEGIN:END:STEP]

# RULE:3: If step value is -ve then it takes from backward direction provided by BEGIN INDEX>END INDEX otherwise we get space or '' as a result.

p="peacock"
print(p[::-1])
print("---------")
print(p[0:2:-1])
print("---------")
print(p[2:-6:-1])
print("---------")
print(p[3:4:-1])
print("---------")
```

```python
#-------SYNTAX-5-------
# strobj=[BEGIN:END:STEP]

# RULE:4: In forward direction if end value is 0 then we get space or '' as a result

s="saranya"
print(s[:0:1])
print("---------")
print(s[1:0:1])
print("---------")
```

```python
# RULE:5: In backward direction if end value is -1 then we get space or '' as a result"

t="saran"
print(t[:-1:-1])
print("---------")
```
